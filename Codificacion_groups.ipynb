## Variables de diagnostico


```python
subcapitulos = data['Dx Principal Cod.1'].unique()
subcapitulos
```

```python
def is_in_range(code, range_str):
    # Eliminar el último carácter si el código tiene 4 caracteres
    if code == '0':
        return False
    if len(code) == 4:
        code = code[:-1]
        
    lower_bound, upper_bound = range_str.split('-')

    # Convertir los códigos alfabéticos en números equivalentes
    code_prefix = ord(code[0])
    lower_bound_prefix = ord(lower_bound[0])
    upper_bound_prefix = ord(upper_bound[0])

    # Manejar el caso de 'X' al final del código
    if 'X' == code[-1]:
        code = code.replace('X', '0')
        code_value_min = int(code[1:])
        code = code.replace('0', '9')
        code_value_max = int(code[1:])
    else:
        code_value_min = code_value_max = int(code[1:])
    
    lower_bound_value = int(lower_bound[1:])
    upper_bound_value = int(upper_bound[1:])
    # Verificar si el código está dentro de los sub-rangos correspondientes a cada letra
    return (
        (lower_bound_prefix < code_prefix < upper_bound_prefix) or
        (lower_bound_prefix == code_prefix == upper_bound_prefix and
         lower_bound_value <= code_value_max and
         code_value_min <= upper_bound_value) or
        (code_prefix == lower_bound_prefix and
         code_value_min >= lower_bound_value) or
        (code_prefix == upper_bound_prefix and
         code_value_max <= upper_bound_value)
    )
```

```python
def is_in_ranges(code, ranges):
    for range_str in ranges:
        if is_in_range(code, range_str):
            return range_str
    return None
```

```python
data.drop(columns=['Dx Principal Cod'], inplace = True)
```

```python
data['Dx Relacionado1 Cod.1'] = np.array([is_in_ranges(x,subcapitulos) for x in data[['Dx Relacionado1 Cod']].values.flatten()])
```

```python
data['Dx Relacionado2 Cod.1 '] = np.array([is_in_ranges(x,subcapitulos) for x in data[['Dx Relacionado2 Cod ']].values.flatten()])
```

```python
data['Dx Relacionado3 Cod.1'] = np.array([is_in_ranges(x,subcapitulos) for x in data[['Dx Relacionado3 Cod']].values.flatten()])
```

```python
regex = r'H[0-5][0-9]'
indices_filtrados = data['Dx Relacionado1 Cod'].str.contains(regex)
data.loc[indices_filtrados, 'Dx Relacionado1 Cod.1'] = 'H00-H59'
indices_filtrados = data['Dx Relacionado2 Cod '].str.contains(regex)
data.loc[indices_filtrados, 'Dx Relacionado2 Cod.1 '] = 'H00-H59'
indices_filtrados = data['Dx Relacionado3 Cod'].str.contains(regex)
data.loc[indices_filtrados, 'Dx Relacionado3 Cod.1'] = 'H00-H59'
```

```python
data.drop(columns=['Dx Relacionado1 Cod','Dx Relacionado2 Cod ','Dx Relacionado3 Cod'], inplace=True)
```

```python
data
```

```python
one_hot_df = pd.DataFrame(columns=subcapitulos)
one_hot_df
```

```python
columns_to_code = ['Dx Relacionado1 Cod.1','Dx Relacionado2 Cod.1 ','Dx Relacionado3 Cod.1','Dx Principal Cod.1']
for index, fila in data.iterrows():
    for valor in subcapitulos:
        if any(fila[columna] == valor for columna in columns_to_code):
            one_hot_df.loc[index, valor] = 1
        else:
            one_hot_df.loc[index, valor] = 0
```

```python
one_hot_df
```

```python
data = pd.concat([data, one_hot_df], axis=1)
```

```python
data.drop(columns=columns_to_code, inplace=True)
data
```

Como hay categorias de diagnosticos comunes entre el principal y relacionados, en este caso se usa un enfoque similar al one-hot encoding, se crea una codificación especial donde se tenga una columna por cada código de diagnóstico posible, y si un paciente tiene ese diagnóstico como principal, el valor sea 2; si es un diagnóstico relacionado, el valor sea 1; y si el paciente no tiene ese diagnóstico, el valor sea 0.
